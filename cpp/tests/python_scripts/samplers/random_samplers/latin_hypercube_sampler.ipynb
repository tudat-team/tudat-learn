```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_lhs(samples):
  fig = plt.figure(figsize=(8,6), facecolor='white')
  ax = plt.gca( )

  tick_font_size = 20

  for i in range(1, samples[0].shape[0]):
    if i == 1:
      plt.plot([-5, 5],[5/samples[0].shape[0] * i, 5/samples[0].shape[0] * i], 'k', linewidth=0.2, label="LHS Bucket Boundaries")
    else:
      plt.plot([-5, 5],[5/samples[0].shape[0] * i, 5/samples[0].shape[0] * i], 'k', linewidth=0.2)
    plt.plot([-5 + 10/samples[0].shape[0] * i, -5 + 10/samples[0].shape[0] * i],[0, 5], 'k', linewidth=0.2)
    print(5/len(samples) * i)

  for i, sample in enumerate(samples):
    plt.scatter(sample[:, 0], sample[:, 1], s=50, label="Latin Hypercube " + str(i + 1))

  plt.xlim(-5,5)
  plt.ylim(0, 5)

  ax.xaxis.set_tick_params(labelsize=tick_font_size)
  ax.yaxis.set_tick_params(labelsize=tick_font_size)
  ax.set_xlabel('x', fontsize=tick_font_size)
  ax.set_ylabel('y', fontsize=tick_font_size)
  ax.legend(fontsize=tick_font_size, loc='lower right', bbox_to_anchor=(1.73, 0.3))

  plt.show()
  
```

```python
three_samples = [
  np.array([
    [-4.088965, 1.371187],
    [ 3.121398, 4.902039],
    [ 1.072085, 2.725378]
  ]),
  np.array([
    [-2.152800, 4.430146],
    [ 3.263513, 1.149127],
    [ 1.421649, 1.677857]
  ]),
  np.array([
    [ 4.794277, 4.241272],
    [-3.837904, 2.426201],
    [ 1.042744, 1.032737]
  ])
]

plot_lhs(three_samples)
```
